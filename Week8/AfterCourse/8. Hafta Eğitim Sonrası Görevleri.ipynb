# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
def main_people():
    how_many=int(input("how many people?"))
    for i in range(how_many):
        print(f"{i+1} people")
        name=input("name?")
        surname=input("surname?")
        age=input("age?")
        heigt=float(input("height?"))
        weight=float(input("weight"))
        vki=find_vki(heigt, weight)
        effect=how_effect_vki(vki)
        output=make_output(name,surname,age,heigt,weight,vki,effect)
        with open("bm_index.txt","a") as file:
            file.write(output)
def find_vki(heigt, weight):
    vki= weight/(heigt**2)
    return vki
def how_effect_vki(vki):
    if vki<18.5:
        return "weak"
    elif vki<=25:
        return "normal"
    elif vki<=30:
        return "over weight"
    elif vki<=40:
        return "obese"
    else:
        return "morbid obese"
def make_output(name,surname,age,heigt,weight,vki,effect):
    a1=f"{name}\n"
    a2=f"{surname}\n"
    a3=f"{age}\n"
    a4=f"{heigt}\n"
    a5=f"{weight}\n"
    a6=f"{vki}\n"
    a7=f"{effect}\n"
    a8="\n"
    output=a1+a2+a3+a4+a5+a6+a7+a8
    return output

main_people()


        

    
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import time
s_menu={
    "çilek": 10,
    "muz": 12,
    "kavun": 15,
    "karpuz": 18
    }
def display_menu():
    print("_________YETBAR MENU_________")
    for item in list(s_menu.keys()):
        print(f"{item.upper()}: {s_menu[item]} TL")
def calculate_product_price(product,number):
    assert number>=0, "product quantity cannot be zero or less"
    if product not in list(s_menu.keys()):
        return None
    else:
        return s_menu[product]*number
def buying():
    display_menu()
    print("Please enter the product to buy. Press 0 to exit")
    purchased_products=[]
    number_purchased_products=[]
    last_price=0
    while 1:
        add_product=input("Which product(s) do you want to add?")
        if add_product =="0":
            print("the receipt is being recorded")
            break
        numbers_will_add=int(input("How many do you want to add?"))
        price=calculate_product_price( add_product, numbers_will_add)
        if price==None:
            print("you entered the wrong value, no product added")
        else:
            print(f"purchase fee: {price}. \nPress 'y' to buy. press 'n' to cancel")
            check=input("Do you confirm the purchase?")
            if check=="y":
                print("purchase confirmed")
                last_price+=price
                purchased_products.append(add_product)
                number_purchased_products.append(numbers_will_add)
                print(f"Subtotal: {last_price}")
            elif check=="n":
                print("purchase has been canceled")
    return last_price, purchased_products, number_purchased_products
def main():
    print("Welcome to YETBAR!\nPress 1 to view the menu, press 2 to make a smoothie, press 0 to exit.")
    program=None
    while 1:
        program=int(input("your preference: "))
        if program<0 or program>2:
            print("wrong entry")
        elif program==0:
            print("exiting...")
            break
        elif program==1:
            display_menu()
        elif program==2:
            last_price, purchased_products, number_purchased_products=buying()
            with open("smoothie.txt","a") as file:
                    file.write(str(time.ctime(time.time())))
                    file.write("\n")
                    file.write(','.join(map(str,purchased_products)))
                    file.write("\n")
                    file.write(','.join(map(str,number_purchased_products)))
                    file.write("\n")
                    file.write(f"sum: {last_price}")
                    file.write("\n")
                    file.write("*"*20)
                    file.write("\n")
            check_exit=input("Is there anything else you want to do?(yes-no)")
            if check_exit=="yes":
                continue
            else:
                print("thanks")
                break
main()




    

    
```
