# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle():
    def __init__(self, name, speed,mileage):
        self.name=name
        self.speed=speed
        self.mileage=mileage
    def Show(self):
        print(f"Name={self.name},speed={self.speed}, mileage={self.mileage}")

class Bus(Vehicle):
    def __init__(self, name, speed,mileage,seats):
        super().__init__(name, speed,mileage)
        self.seats=seats
    def Show(self):
        print(f"Name={self.name},speed={self.speed}, mileage={self.mileage}, seats={self.seats}")


Volvo=Bus("school Volvo",180,12,16)
Renault=Vehicle("Renault",150,15)
Volvo.Show()
Renault.Show()
````
